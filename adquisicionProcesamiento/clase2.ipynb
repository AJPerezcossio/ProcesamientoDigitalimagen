```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img= cv2.imread('descarga.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(1, 4, figsize=(20, 5))

axs[0].imshow(img_rgb)
axs[0].set_title('Imagen')

fig_img=axs[1].imshow(img_rgb[..., 0], cmap='gray')
fig.colorbar(fig_img, ax=axs[1],shrink=0.5)
axs[1].set_title('Canal Rojo')

fig_img=axs[2].imshow(img_rgb[..., 1], cmap='gray')
fig.colorbar(fig_img, ax=axs[2],shrink=0.5)
axs[2].set_title('Canal Verde')

fig_img=axs[3].imshow(img_rgb[..., 2], cmap='gray')
fig.colorbar(fig_img, ax=axs[3],shrink=0.5)
axs[3].set_title('Canal Azul')
```

escala de grises 


```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

# Cargar imagen en escala de grises
img = cv2.imread('descarga.jpg', cv2.IMREAD_GRAYSCALE)

# VALORES INICIALES DE PROFUNDIDAD DE COLORES
original_dtype = img.dtype
print('Profundidad de color original:', original_dtype)

# Convertir a 16 bits (aumentar profundidad)
img16bits = np.uint16(img) * 256  # Multiplicar para escalar valores 0-255 a 0-65535
new_dtype = img16bits.dtype
print('Nueva profundidad de color:', new_dtype)

# Guardado con OpenCV (opcional)
cv2.imwrite('imagen_16bits.png', img16bits)

# Visualización con matplotlib
plt.figure(figsize=(15, 7))

plt.subplot(1, 2, 1)
plt.title('Imagen Original (8 bits)')
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.colorbar(shrink=0.5)

plt.subplot(1, 2, 2)
plt.title('Imagen 16 bits')
plt.imshow(img16bits, cmap='gray', vmin=0, vmax=65535)
plt.colorbar(shrink=0.5)

plt.tight_layout()
plt.show()

# Mostrar información adicional
print(f"\nDimensiones de la imagen: {img.shape}")
print(f"Valor mínimo en original: {img.min()}")
print(f"Valor máximo en original: {img.max()}")
print(f"Valor mínimo en 16 bits: {img16bits.min()}")
print(f"Valor máximo en 16 bits: {img16bits.max()}")
```
