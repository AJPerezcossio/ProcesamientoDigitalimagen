cargar imagenes 


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


img_bgr = cv2.imread('foto1.jpeg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gris = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)


cv2.imwrite('foto1_gris.jpg', img_gris)
cv2.imwrite('foto1_bgr.jpg', img_bgr)

print("foto1.jpeg cargado y procesado")


plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.imshow(img_rgb)
plt.title('Original (RGB)')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img_gris, cmap='gray')
plt.title('Escala de Grises')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(img_rgb[..., 0], cmap='gray')
plt.title('Canal Rojo (Gris)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(img_rgb[..., 0], cmap='Reds')
plt.title('Canal Rojo (Color Falso)')
plt.axis('off')

plt.suptitle('Análisis de presi - Modo 1', fontsize=16)
plt.show()
```
