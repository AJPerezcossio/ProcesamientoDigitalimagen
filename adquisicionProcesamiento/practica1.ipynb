```python
import cv2  
import matplotlib.pyplot as plt
img= cv2.imread('descarga.jpg') 
plt.rcParams['figure.figsize'] = (10,5)
img= cv2.imread('descarga.jpg',cv2.IMREAD_COLOR)

#separar por coores RGB
plt.subplot(1,4,1)
plt.title('imagen original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(1,4,2)
plt.title('CANAL ROJO')
plt.imshow(img[:,:,0],cmap='gray')

plt.subplot(1,4,3)
plt.title('CANAL VERDE')
plt.imshow(img[:,:,1],cmap='gray')

plt.subplot(1,4,4)
plt.title('CANAL AZUL')
plt.imshow(img[:,:,2],cmap='Blues')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (10,5)
img= cv2.imread('descarga.jpg',cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r=imgRGB[:,:,2]
g=imgRGB[:,:,1]
b=imgRGB[:,:,0]

#CREAR IMAGENES POR CANALES 

r_img=np.zeros_like(imgRGB)
r_img[:,:,0]=r

g_img=np.zeros_like(imgRGB)
g_img[:,:,1]=g

b_img=np.zeros_like(imgRGB)
b_img[:,:,2]=b

plt.subplot(141)
plt.title('imagen original')
plt.imshow(imgRGB)

plt.subplot(142)
plt.title('CANAL ROJO')
plt.imshow(r_img)

plt.subplot(143)
plt.title('CANAL VERDE')
plt.imshow(g_img)

plt.subplot(144)
plt.title('CANAL AZUL')
plt.imshow(b_img)

plt.show()
```
